```python
def ratio_user_score(game):
  game.plot.scatter(x = 'user_score', y = 'sum_sales', alpha=0.3)
  plt.title('–ó–∞–≤–∏—Å–∏–º–æ—Å—Ç—å –ø—Ä–æ–¥–∞–∂ –æ—Ç –æ—Ç–∑—ã–≤–æ–≤ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π')
  plt.xlabel('–†–µ–π—Ç–∏–Ω–≥ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π')
  plt.ylabel('–ü—Ä–æ–¥–∞–∂–∏')
  corr = round(game['user_score'].corr(game['sum_sales']), 2)
  result = '–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏: ' + str(corr)
  return result

def ratio_critic_score(game):
  game.plot.scatter(x = 'critic_score', y = 'sum_sales', alpha=0.3)
  plt.title('–ó–∞–≤–∏—Å–∏–º–æ—Å—Ç—å –ø—Ä–æ–¥–∞–∂ –æ—Ç –æ—Ç–∑—ã–≤–æ–≤ –∫—Ä–∏—Ç–∏–∫–æ–≤')
  plt.xlabel('–†–µ–π—Ç–∏–Ω–≥ –∫—Ä–∏—Ç–∏–∫–æ–≤')
  plt.ylabel('–ü—Ä–æ–¥–∞–∂–∏')
  corr = round(game['critic_score'].corr(game['sum_sales']), 2)
  result = '–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏: ' + str(corr)
  return result
```

```python
# –í—ã–±–æ—Ä —Ç–æ–ø-5 –ø–ª–∞—Ç—Ñ–æ—Ä–º
top_platforms = platform_sales[:5]
# –°—É–º–º–∏—Ä–æ–≤–∞–Ω–∏–µ –ø—Ä–æ–¥–∞–∂ –æ—Å—Ç–∞–ª—å–Ω—ã—Ö –ø–ª–∞—Ç—Ñ–æ—Ä–º
other_sales = platform_sales[5:].sum()
# –î–æ–±–∞–≤–ª–µ–Ω–∏–µ –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ "–¥—Ä—É–≥–∏–µ" –≤ —Å–ø–∏—Å–æ–∫ —Ç–æ–ø-5
top_platforms.loc[len(top_platforms.index )] = other_sales
display(type(top_platforms))
```

```python
def popular_platforms_multiple():
    regions = [('na_sales', '–°–µ–≤–µ—Ä–Ω–æ–π –ê–º–µ—Ä–∏–∫–µ'), ('eu_sales', '–ï–≤—Ä–æ–ø–µ'), ('jp_sales', '–Ø–ø–æ–Ω–∏–∏')]
    fig, axes = plt.subplots(1, 3, figsize=(18, 6)) 
    fig.suptitle('–û–±–∑–æ—Ä —Ä—ã–Ω–∫–∞ –ø–ª–∞—Ç—Ñ–æ—Ä–º') 
    for i, (region, name) in enumerate(regions):
        sales_t = actual_games.pivot_table(index='platform', values=region, aggfunc='sum')\
                                     .sort_values(ascending=False, by=region)
        platform_sales = sales_t[:5]
        platform_sales.loc[len(platform_sales.index)] = sales_t[5:].sum()
        platform_sales.rename(index={5:"other"}, inplace= True)
        platform_sales.plot(kind='pie', y=region, autopct='%1.1f%%', 
                            colors=['#7169c9', '#a3a0fb', '#cfc3ff', '#e8e5fd', '#f0f0f0'],
                            ax=axes[i], legend=False, title='–ü–æ–ø—É–ª—è—Ä–Ω–æ—Å—Ç—å –ø–ª–∞—Ç—Ñ–æ—Ä–º –≤ ' + name)
        axes[i].set_ylabel('')
    plt.show()
```

```python
popular_platforms_multiple()
```

    <font size="3", color = "black">
<br /> –ö—Ä—É–≥–æ–≤–∞—è –¥–∏–∞–≥—Ä–∞–º–º–∞ –∏–∑ –¢–û–ü-5 –∏—Å–∫–∞–∂–∞–µ—Ç –¥–æ–ª–∏ —Ä—ã–Ω–∫–∞, –∫–æ—Ç–æ—Ä—ã–µ –∑–∞–Ω–∏–º–∞—é—Ç –ø–µ—Ä–≤—ã–µ —Ç–æ–ø-5 –ø–ª–∞—Ç—Ñ–æ—Ä–º, —Ç.–∫. –∏—Ö —Å—É–º–º–∞ –ø—Ä–æ–¥–∞–∂ –¥–µ–ª–∏—Ç—Å—è –Ω–∞ 100%, –Ω–æ —É –Ω–∞—Å –∏–º–µ—é—Ç—Å—è –¥—Ä—É–≥–∏–µ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã, –∫–æ—Ç–æ—Ä—ã–µ —Ç–∞–∫–∂–µ –∑–∞–Ω–∏–º–∞—é—Ç –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω—É—é –¥–æ–ª—é —Ä—ã–Ω–∫–∞

```python
def popular_genre_multiple():
    regions = [('na_sales', '–°–µ–≤–µ—Ä–Ω–æ–π –ê–º–µ—Ä–∏–∫–µ'), ('eu_sales', '–ï–≤—Ä–æ–ø–µ'), ('jp_sales', '–Ø–ø–æ–Ω–∏–∏')]
    fig, axes = plt.subplots(1, 3, figsize=(18, 6)) 
    fig.suptitle('–û–±–∑–æ—Ä –ø–æ–ø—É–ª—è—Ä–Ω—ã—Ö –∂–∞–Ω—Ä–æ–≤') 
    for i, (region, name) in enumerate(regions):
       sales_t = actual_games.pivot_table(index='genre', values=region, aggfunc='sum')\
                                     .sort_values(ascending=False, by=region)
       platform_sales = sales_t[:5]
       platform_sales.loc[len(platform_sales.index)] = sales_t[5:].sum()
       platform_sales.rename(index={5:"other"}, inplace= True)
       platform_sales.plot(kind='pie', y=region, autopct='%1.1f%%', 
                           colors=['#7169c9', '#a3a0fb', '#cfc3ff', '#e8e5fd', '#f0f0f0'],
                           ax=axes[i], legend=False, title='–ü–æ–ø—É–ª—è—Ä–Ω–æ—Å—Ç—å –∂–∞–Ω—Ä–æ–≤ –≤ ' + name)
       axes[i].set_ylabel('')
    plt.show()
```

```python
popular_genre_multiple()
```

–í –°–µ–≤–µ—Ä–Ω–æ–π –ê–º–µ—Ä–∏–∫–µ —Å–∞–º—ã–µ –ø–æ–ø—É–ª—è—Ä–Ω—ã–µ –∂–∞–Ω—Ä—ã: action, shooter, sports, role-playing –∏ misc;

```python
def popular_rating_multiple():
    regions = [('na_sales', '–°–µ–≤–µ—Ä–Ω–æ–π –ê–º–µ—Ä–∏–∫–µ'), ('eu_sales', '–ï–≤—Ä–æ–ø–µ'), ('jp_sales', '–Ø–ø–æ–Ω–∏–∏')]
    fig, axes = plt.subplots(1, 3, figsize=(18, 6)) 
    fig.suptitle('–û–±–∑–æ—Ä –ø–æ–ø—É–ª—è—Ä–Ω—ã—Ö —Ä–µ–π—Ç–∏–Ω–≥–æ–≤') 
    for i, (region, name) in enumerate(regions):
        platform_sales = actual_games.pivot_table(index='rating', values=region, aggfunc='sum')\
                                     .sort_values(ascending=False, by=region)[:5]
        platform_sales.plot(kind='pie', y=region, autopct='%1.1f%%', 
                            colors=['#7169c9', '#a3a0fb', '#cfc3ff', '#e8e5fd', '#f0f0f0'],
                            ax=axes[i], legend=False, title='–ü–æ–ø—É–ª—è—Ä–Ω–æ—Å—Ç—å —Ä–µ–π—Ç–∏–Ω–≥–æ–≤ –≤ ' + name)
        axes[i].set_ylabel('')
    plt.show()
```
